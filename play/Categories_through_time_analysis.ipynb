## Detrend by appearance of words in the OED

```python
#load word data
project_data_path = "/Users/mark/Dropbox/Math/Complex_Systems/research/classifying-ideas/historical_wikipedia/data/"
with open(project_data_path + "words_with_years.json", "r") as f:
    words_with_years = json.load(f)
```

```python
len(words_with_years)
```

```python
#turn into dataframe
words_df = pd.DataFrame.from_dict(words_with_years, orient='index')
words_df.columns = ['year']
```

```python
words_df.head()
```

```python
words_df[words_df['year'] > 1890].shape[0]
```

```python
def get_num_words(year):
    """
    returns the number of words in the OED 
    based on given year
    """
    return words_df[words_df['year'] > year].shape[0]
```

```python
#add words by year to categories dataframe
decades = range(1000, 2020, 10)
words_by_year_series = pd.Series(map(get_num_words, decades), index=decades)
words_by_year_series
```

```python
categories_df['A language']
```

```python
#match data type of index (change to int64)

categories_df.index = categories_df.index.astype(dtype="int64")
categories_df.index
```

```python
categories_detrended_df = categories_df.subtract(words_by_year_series, axis='index')
```

```python
categories_detrended_df.plot()
```
